```python
import sys 
sys.path.append('../..')

import torch 
import torch_scatter 

import pyg_util 
```

```python
num_nodes = 5
num_edges = 15 

node_feat = torch.randn(num_nodes, 64)
edge_index = torch.randint(low=0, high=num_nodes, size=(2, num_edges))
```

```python
out_1 = torch_scatter.scatter_add(
    src = node_feat[edge_index[0]],
    index = edge_index[1],
    dim = 0,
    dim_size = num_nodes,
)

out_1.shape 
```

```python
edge_index = pyg_util.sort_edge_index(edge_index, num_nodes, 'col')

out_2 = pyg_util.scatter_add(
    src = node_feat[edge_index[0]],
    index = edge_index[1],
    size = num_nodes,
)

out_2.shape 
```

```python
(out_2 - out_1).abs().sum() 
```
